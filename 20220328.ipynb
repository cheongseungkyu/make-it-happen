```python
def test(doname):
  URL = ('http://apis.data.go.kr/1400377/forestPoint/'
      'forestPointListSidoSearch?'
      'serviceKey=INgUFCPLVXKA1R1IoOaejy9GNI%2BuFuWNZPPInGnjgr2XQQYmmq09MvpkvnykS3Hp5pd1TD1ysdH%2F1M5k9GghmQ%3D%3D'
      '&numOfRows=10'
      '&pageNo=1'
      '&doname='+doname+
      '&excludeForecast=0'
      '&_type=json')

  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']['item']

  doname = [item['doname'] for item in items]
  analdate = [item['analdate'] for item in items]
  d1 = [item['d1'] for item in items]
  d2 = [item['d2'] for item in items]
  d3 = [item['d3'] for item in items]
  d4 = [item['d4'] for item in items]
  meanavg = [item['meanavg'] for item in items]

  doname = pd.Series(doname)
  analdate = pd.Series(analdate)
  d1 = pd.Series(d1)
  d2 = pd.Series(d2)
  d3 = pd.Series(d3)
  d4 = pd.Series(d4)
  meanavg = pd.Series(meanavg)

  df = pd.DataFrame({'doname' : doname, 'analdate' : analdate, 'd1' : d1, 'd2' : d2, 'd3' : d3, 'd4' : d4, 'meanavg' : meanavg})
  return df
```

```python
iface = gd.Interface(test, gd.inputs.Dropdown(['서울특별시', '부산광역시','대구광역시','경기도','강원도','대전광역시','인천광역시','광주광역시','울산광역시','세종특별자치시','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주특별자치도']), outputs=['dataframe'])
iface.launch(debug=True)
```
