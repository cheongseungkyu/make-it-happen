```python
import pandas as pd

items = r['response']['body']['items']['item']

analdate = [item['analdate'] for item in items]
doname = [item['doname'] for item in items]
d1 = [item['d1'] for item in items]
d2 = [item['d2'] for item in items]
d3 = [item['d3'] for item in items]
d4 = [item['d4'] for item in items]
meanavg = [item['meanavg'] for item in items]

analdate = pd.Series(analdate)
doname = pd.Series(doname)
d1 = pd.Series(d1)
d2 = pd.Series(d2)
d3 = pd.Series(d3)
d4 = pd.Series(d4)
meanavg = pd.Series(meanavg)

df = pd.DataFrame({'doname' : doname, 'analdate' : analdate, 'd1' : d1, 'd2' : d2, 'd3' : d3, 'd4' : d4, 'meanavg' : meanavg})
```

doname : 산불위험지역 시도 명칭
analdate : 데이터기준일자
d1 : 산불위험예보지수 낮음
d2 : 산불위험예보지수 다소높음
d3:  산불위험예보지수 높음
d4:  산불위험예보지수 매우높음
meanavg : 산불위험예보지수 평균
